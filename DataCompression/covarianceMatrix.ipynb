```python
import seaborn as sns

import pandas as pd
pd.options.display.float_format = "{:,.2f}".format

from sklearn import datasets
import numpy as np

data = datasets.load_iris()

df = pd.DataFrame(data=np.hstack((data.data, data.target[:, np.newaxis])), columns=[*data.feature_names, 'label'])
labels_map = {key: value for key, value in enumerate(data.target_names)}
df['label'] = df['label'].map(labels_map)
sns.pairplot(df, hue="label", palette={data.target_names[0]: 'r', data.target_names[1]: 'g', data.target_names[2]: 'b'})
```

```python
from sklearn.preprocessing import StandardScaler

stdsc = StandardScaler()
X_std = stdsc.fit_transform(data.data)

covariance_matrix = np.cov(X_std.T)

df_cov = pd.DataFrame(covariance_matrix)
df_cov
```
